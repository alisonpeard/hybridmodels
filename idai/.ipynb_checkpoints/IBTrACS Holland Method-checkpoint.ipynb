Storm of interest: Hurrican Katrina, New Orleans, 2005

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from os.path import join
import folium
```

```python
hurricane = "KATRINA"
season = 2005

global_data = join("~", "DPhil archive", "data", "global")
ibtracs = pd.read_csv(join(global_data, "ibtracs_formatted_yu.csv"))
ibtracs = ibtracs[(ibtracs.NAME==hurricane) & (ibtracs.SEASON==season)]
ibtracs = ibtracs.reset_index()
ibtracs = gpd.GeoDataFrame(ibtracs, geometry=gpd.points_from_xy(ibtracs.LON, ibtracs.LAT))
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# We restrict to South America.
ax = world[world.continent==''].plot(
    color='white', edgecolor='black')

ibtracs.plot(ax=ax)
```

```python
def wind_field(vm, r, rm=10, f=5e-5):
    """
    Parameters:
    f : float
        Coriolis parameter: based-on Emmanual and Rotunno (2011)
    """
    v = 2 * r * (rm * vm - 0.5 * f * rm ** 2) / (rm ** 2 + r ** 2) - 0.5 * f * r
    return v
    
```

```python
rmax = 1000

for r in range(rmax):
    print(wind_field(ibtracs.WMO_WIND[0], r, rm=rmax))
          
```
