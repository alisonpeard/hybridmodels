Need multiple files for AOI so need to mosaic and clip them.

```python
fabdemfile = join(save_dir, "elevation.tif")
                  
# mosaic the files in fabdemfiles
mosaic, transform = merge(fabdemfiles)

with rasterio.open(fabdemfiles[0]) as src:
    out_meta = src.meta.copy()
    out_crs = src.crs

out_meta.update({"driver": "GTiff",
                 "height": mosaic.shape[1],
                 "width": mosaic.shape[2],
                 "transform": transform,
                 "crs": out_crs
                })

with rasterio.open(fabdemfile, "w", **out_meta) as dest:
    dest.write(mosaic)

# crop the mosaic
with rasterio.open(fabdemfile) as src:
    out_meta = src.meta.copy()
    out_crs = src.crs
    cropped, transform = mask(src, [aoi_lonlat], crop=True)

out_meta.update({"driver": "GTiff",
                 "height": cropped.shape[1],
                 "width": cropped.shape[2],
                 "transform": transform,
                 "crs": out_crs
                })

with rasterio.open(fabdemfile, "w", **out_meta) as dest:
    dest.write(cropped)
```

```python
# save Google files to local machine then load using rasterio...
features = ["elevation", "aspect", "curvature", "slope", "soilcarbon", "mangroves", "ndvi_mfmask", "evi_mfmask", "jrc_permwater", "precipitation"]
feature_gdf = grid_lonlat.copy(deep=True)

def get_zonal_stats(feature, grid, aoi, save_dir, statistic="mean"):
    """EPSG:4326"""
    # crop and calculate mean zonal statistics
    with rasterio.open(join(save_dir, f"{feature}.tif")) as src:
        out_meta = src.meta.copy()
        out_crs = src.crs
        ndval = src.nodatavals[0]
        cropped, affine = mask(src, [aoi], crop=True)
        cropped = cropped[0, :, :].astype('float64')
        cropped[cropped==ndval] = np.nan

    feature_list = rasterstats.zonal_stats(grid, cropped, affine=affine, geojson_out=True, nodata=np.nan)    

    return feature_list


for feature in features:
    # crop and calculate mean zonal statistics
    feature_list = get_zonal_stats(feature, grid_lonlat, aoi_lonlat, save_dir)
    feature_gdf[f"{feature}"] = [feature_list[i].get("properties").get("mean") for i in range(len(feature_list))]

feature_gdf.to_file(join(save_dir, "feature_stats.shp"))
```

## Distance and slope to permanent water
For now, use QGIs to fix geometries, and extract values >= 10 as permanent water layer for conservative estimate. Save these as *jrc_permwater.shp* files.

Need to change to a projection that uses meters for accurate slope. Create a output geopandas dataframe for all geomorphology features.

For slope, currently using mean elevation at the grid cell over distance to permanent water. Later, it would be better to calculate the average slope over the linestring connecting said grid cell to permanent water.

```python
from shapely.ops import nearest_points
from rasterio.sample import sample_gen

def dist_slope_to_water(df, water, elev, geom_col='geometry'):
    """Function adapted from Muckley (2020) for JRC permanent water dataset.
    
    NB: make sure all geodataframes in meters projection.
    To do: check slope calculations. Too simple.
    
    References:
    -----------
    ..[1] https://github.com/leomuckley/Multi-Input-ConvLSTM/blob/master/utils/feature_generation.py
    """
    from shapely.ops import nearest_points
    water_union = water.unary_union  # combine into one polygon
    
    dist_list =[]
    slope_list = []
    
    for index, row in df.iterrows():
        nearest = nearest_points(row[geom_col], water_union)[1]
        dist = row[geom_col].distance(nearest)
        dist_list.append(dist)
        
        higher = elev.iloc[index, :]["elevation"]
        slope = higher / dist if dist > 0 else np.nan # assumption that permanent water is at sea-level
        slope_list.append(slope)

    return dist_list, slope_list
```

```python
# switch to Pseudo-Mercator for calculations in meters
feature_gdf_pm = feature_gdf.to_crs("EPSG:3857")
pw_gdf_pm = gpd.read_file(join(save_dir, "jrc_permwater.shp")).to_crs("EPSG:3857")

dist_list, slope_list = dist_slope_to_water(grid, pw_gdf_pm, feature_gdf_pm)

feature_gdf["dist_pw"] = dist_list
feature_gdf["slope_pw"] = slope_list

feature_gdf.to_file(join("..", "beira", "outdata", "feature_stats.shp"))
feature_gdf.head()
```
